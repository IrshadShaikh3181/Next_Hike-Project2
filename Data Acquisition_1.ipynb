```python
# Convert date column to datetime format
dataset_1 = pd.read_excel('dataset_1.xlsx')  
dataset_1['dteday'] = pd.to_datetime(dataset_1['dteday'], errors='coerce')

# Sort the dataset by date
dataset_1.sort_values('dteday', inplace=True)
```

```python
print(dataset_1)
```

```python
# Grouping by season and calculating average temperature
season_avg_temp = dataset_1.groupby('season')['temp'].mean()

# Bar plot
season_avg_temp.plot(kind='bar', color='orange', figsize=(8, 5))
plt.title('Average Temperature by Season')
plt.xlabel('Season')
plt.ylabel('Normalized Temperature')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'], rotation=0)
plt.grid(axis='y')
plt.show()
```

#Insights:The data is of Spring seasons and average temperature lies between 0 to 0.2 degrees celsius.

```python
import matplotlib.pyplot as plt

# Line plot for temperature trends
plt.figure(figsize=(12, 6))
plt.plot(dataset_1['dteday'], dataset_1['temp'], color='blue', label='Temperature')
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Normalized Temperature')
plt.legend()
plt.grid()
plt.show()
```

#Insights:Temperatue is low between 21st to 25th of January

```python
# Calculate 7-day rolling average for temperature
dataset_1['rolling_temp'] = dataset_1['temp'].rolling(window=7).mean()

# Plot rolling average
plt.figure(figsize=(10, 6))
plt.plot(dataset_1['dteday'], dataset_1['temp'], color='blue', label='Original Temperature')
plt.plot(dataset_1['dteday'], dataset_1['rolling_temp'], color='red', label='7-Day Rolling Average')
plt.title('Smoothed Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Normalized Temperature')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def univariate_analysis(data):
    for col in data.columns:
        # If the column is representing a category
        if data[col].dtypes == "object":
            plt.figure(figsize=[12, 6])
            # This is for the first subplot
            plt.subplot(1, 2, 1)
            sns.barplot(x=data[col].value_counts().index, 
                        y=data[col].value_counts().values, color="darkblue")
            plt.title("Categorical Distribution of {}".format(col))
            # This is for the second subplot
            plt.subplot(1, 2, 2)
            plt.pie(x=data[col].value_counts().values, 
                    labels=data[col].value_counts().index, autopct='%1.2f%%')
            plt.title("Categorical Distribution of {}".format(col))
            plt.show()
        # If the column is representing a numerical field
        elif data[col].dtypes in ["int64", "float64"]:
            plt.figure(figsize=[12, 6])
            # This is for the first subplot
            plt.subplot(1, 2, 1)
            sns.histplot(x=col, data=data, color="darkblue")
            plt.title("Continuous Distribution of {}".format(col))
            # This is for the second subplot
            plt.subplot(1, 2, 2)
            sns.boxplot(x=col, data=data, color="darkblue")
            plt.title("Continuous Distribution of {}".format(col))
            plt.show()
```

```python
univariate_analysis(dataset_1)
```

```python
# Count the occurrences of each category in the selected columns
season_counts = dataset_1['season'].value_counts()
weathersit_counts = dataset_1['weathersit'].value_counts()
weekday_counts = dataset_1['weekday'].value_counts()

# Create subplots
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plot first pie chart (season)
ax[0].pie(season_counts, labels=season_counts.index, autopct='%1.1f%%', startangle=140)
ax[0].set_title('Season Distribution')

# Plot second pie chart (weathersit)
ax[1].pie(weathersit_counts, labels=weathersit_counts.index, autopct='%1.1f%%', startangle=140)
ax[1].set_title('Weather Situation Distribution')

# Plot third pie chart (weekday)
ax[2].pie(weekday_counts, labels=weekday_counts.index, autopct='%1.1f%%', startangle=140)
ax[2].set_title('Weekday Distribution')

# Show plots
plt.tight_layout()
plt.show()
```

```python
categorical_cols, numerical_cols = [], []

for i in dataset_1.columns:
    print("The datatype for {} is {}".format(i, dataset_1[i].dtypes))
    
    # Check if the datatype is object (categorical)
    if dataset_1[i].dtypes == "object":
        # If the datatype is object: fillna with mode of that column
        mode_value = dataset_1[i].mode()[0]
        dataset_1[i].fillna(mode_value, inplace=True)
        categorical_cols.append(i)
    else:
        # If the datatype is not an object (numerical)
        # Check if the datatype is boolean
        if dataset_1[i].dtypes == "bool":
            dataset_1[i].fillna(False, inplace=True)
        else:
            # Else if the datatype is numerical: fillna with mean of that column
            mean_value = dataset_1[i].mean()
            dataset_1[i].fillna(mean_value, inplace=True)
        numerical_cols.append(i)

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example: Create a violin plot with 'hum' bins as hue
plt.figure(figsize=(8, 6))
sns.violinplot(
    x=pd.cut(dataset_2['hum'], bins=5),
    y='cnt',
    data=dataset_2,
    hue=pd.cut(dataset_2['hum'], bins=5),  # Assigning hue
    palette='coolwarm',
    inner='quartile',
    legend=False
)
plt.title('Violin Plot: Humidity vs Total Rentals')
plt.xlabel('Humidity Bins')
plt.ylabel('Total Rentals')
plt.xticks(rotation=45)
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Binning the 'hum' column into categories
dataset_2['hum_bins'] = pd.cut(dataset_2['hum'], bins=5, labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High'])

# Boxplot: Compare 'hum_bins' (binned humidity) across 'cnt' (total rentals)
plt.figure(figsize=(8, 5))
sns.boxplot(x='hum_bins', y='cnt', data=dataset_2, palette='Set3')
plt.title('Humidity vs Total Rentals (Binned)')
plt.xlabel('Humidity Levels')
plt.ylabel('Total Rentals')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x=pd.cut(dataset_2['windspeed'], bins=5), y='cnt', data=dataset_2, palette='coolwarm', inner='quartile')
plt.title('Wind Speed vs Total Rentals')
plt.xlabel('Wind Speed Bins')
plt.ylabel('Total Rentals')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Summing up the counts of 'casual' and 'registered'
rental_types = ['Casual', 'Registered']
rental_counts = [dataset_2['casual'].sum(), dataset_2['registered'].sum()]

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(rental_counts, labels=rental_types, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Proportion of Rental Types')
plt.show()
```

#Count of registered rentals is more as compared to casuals

```python
# Create wind speed bins
dataset_2['windspeed_bins'] = pd.cut(dataset_2['windspeed'], bins=5, labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High'])

# Group by wind speed bins and sum up total rentals
windspeed_groups = dataset_2.groupby('windspeed_bins')['cnt'].sum()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(windspeed_groups, labels=windspeed_groups.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightpink', 'lightblue', 'orange', 'purple'])
plt.title('Total Rentals by Wind Speed Levels')
plt.show()
```

#Number of rentals are more if the wind speed is low.
And less if there is high wind speed there might be weather issues.
